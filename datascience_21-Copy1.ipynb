```python
import matplotlib.pyplot as plt
data1=[72,76,24,14,67,52,75,78,31,32]
data2=[62,5,91,25,36,32,96,95,30,90]
data3=[23,89,12,78,72,89,25,69,68,86]
bdata=(data1,data2,data3)
box=plt.boxplot(bdata,vert=1,patch_artist=True)
colors=['green','red','pink']
for p, color in zip(box['boxes'],colors):
    p.set_facecolor(color)
plt.show()
```

```python
import matplotlib.pyplot as plt
umemp_rate=[6.1,5.8,5.7,5.7,5.8,5.6,5.5,5.3,5.2,5.2]
index_price=[1500,1520,1525,1523,1515,1540,1547,1560,1555,1565]
plt.scatter(umemp_rate,index_price,c='b')
plt.xlabel('umemp_rate',fontsize=16)
plt.ylabel('index_price',fontsize=16)
plt.title('Relationship between unemployement Rate and stock index price',fontsize=20)
plt.show()
```
